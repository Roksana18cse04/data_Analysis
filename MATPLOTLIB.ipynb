```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,ax=plt.subplots()  # create a figure containning a single axis
ax.plot([1,2,3,4,5,6,7],[1,7,6,5,4,2,3])   #some data in axix
plt.show()   # figure show
```

```python
fig,ax=plt.subplots()
ax.set_title("my_first plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.plot([1,2,3,4],[1,4,2,3])
ax.grid()   # for graph show
ax.xaxis.set_ticks([1,2,3,4])
ax.yaxis.set_ticks([1,2,3,4])
plt.show()
```

```python
fig = plt.figure()             # an empty figure with no Axes
fig, ax = plt.subplots()       # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one Axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```
