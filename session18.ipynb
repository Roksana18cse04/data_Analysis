```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
df.head()
```

```python
sns.catplot(data=df,x='day',y='total_bill',jitter=False)
```

```python
sns.catplot(data=df,x="day",y="total_bill",hue="sex",kind="swarm")
```

```python
sns.catplot(data=df,x="smoker",y="tip",order=["No","Yes"])
```

```python
sns.catplot(data=df,x="day",y="total_bill",hue="time",kind="swarm")
```

```python
sns.catplot(data=df,x='day',y='total_bill',kind="box")
```

```python
sns.catplot(data=df,x='day',y='total_bill',jitter=True)
```

```python
sns.catplot(data=df,x='day',y='total_bill',hue="smoker",kind="box")
```

```python
sns.catplot(data=df,x='day',y='total_bill',hue="smoker",kind="swarm",col="time",aspect=.7)
```

```python
diamonds=sns.load_dataset('diamonds')
sns.catplot(
    data=diamonds.sort_values("color"),
    x="color",y="price",kind="boxen",
)
```

```python
sns.catplot(data=df, x='total_bill',y='day',hue="sex",kind="violin")
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
titanic.head()
```

```python
sns.catplot(data=titanic,x="age",y="deck",errorbar=("pi",95),kind="bar")
```

```python
sns.catplot(data=titanic,x="age",y="deck",errorbar=("sd",95),kind="bar")
```

```python
sns.catplot(data=titanic,x="sex",y="survived",hue="class",kind="point")
```

```python
penguins=sns.load_dataset("penguins")
penguins
```

```python
sns.displot(penguins,x="flipper_length_mm")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species",kind="kde",multiple="stack")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species",kind="kde",multiple="stack",fill=False)
```

```python
sns.pairplot(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot ,fill=True)
g.map_diag(sns.histplot, kde=True)
```
